```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Base Sangre 2.csv")
df.head()
```

```python
# To convert categorical variables to dummy variables

col_categ = ["Sexo", "Tipo", "T", "N", "M", "TNM", "Borrman"]

data_categ = pd.get_dummies(df[col_categ])
data_categ
```

```python
# concatening numerical and dummy variables

df_t = (pd.concat([df.drop(col_categ, axis=1), data_categ], axis=1))
df_t
```

```python
# Dataframe was standarized

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df_normalizado = pd.DataFrame(StandardScaler().fit_transform(df_t))
```

```python
#Data clustering of standarized dataframe

from sklearn.cluster import MiniBatchKMeans

estimador = MiniBatchKMeans(n_clusters=2)
estimador.fit(df_normalizado)
etiquetas_pred = estimador.labels_
```

```python
# Showing predicted clustering labels

etiquetas_pred
```

```python
# That function allow us to resume and append clinical characteristics of the clusters

def resumen_cluster(cluster_id):
    cluster = df_t[etiquetas_pred==cluster_id]
    resumen_cluster = cluster.mean().round().to_dict()
    resumen_cluster["cluster_id"] = cluster_id
    return resumen_cluster

def comparar_clusters(*cluster_ids):
    resumenes = []
    for cluster_id in cluster_ids:
        resumenes.append(resumen_cluster(cluster_id))
    return pd.DataFrame(resumenes).set_index("cluster_id").T
```

```python
#Showing clinical characteristics of each cluster

comparar_clusters(0,1)
```

```python
# 

def resumen_cluster(cluster_id):
    cluster = df[etiquetas_pred==cluster_id]
    resumen_cluster = cluster.mean().to_dict()
    resumen_cluster["cluster_id"] = cluster_id
    return resumen_cluster
comparar_clusters(0,1)
```

```python
# We create a dataframe with the clustering labels for each patient

df["Grupo"] = etiquetas_pred
df
```
